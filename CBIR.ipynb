# Recherche par image entière

Pour une première implémentation d’un système CBIR, l’objectif est d’afficher les 5 images les plus similaires à l’image requête « Imrequest.jpg » en se basant sur un calcul de différence entre images entières.
L'image requete est dans un dossier "img_requetes"

```python
image_requete = image.imread('/content/drive/MyDrive/CBIR/img_requetes/ImageRequete2.jpg')
plt.imshow(image_requete)
```

Nous allons utiliser la distance enclidienne entre l'image requéte et toutes les images de la base. Les distances seront enregistrées dans un dictionnaire et seront par la suite triées par ordre croissant.

```python
# Comparer les pixels des deux images utilisant une distance enculidienne
import scipy.spatial.distance as dist
distances = {}
for i in range(len(loaded_image)):
  distances[i]= dist.euclidean(image_requete.flatten(),loaded_image[i].flatten())
distances = sorted([(v,k) for (k,v) in distances.items() ])

print(distances)
```

```python
#Afficher les 6 premiers
fig = plt.figure(figsize=(6,6),dpi=200)
for i in range(6):
  ax1 = fig.add_subplot(2,3,i+1)
  plt.imshow(loaded_image[distances[i][1]])
  plt.title(imageNames[distances[i][1]])

plt.show()
```

# Recherche par Couleur

```python
# la CBIR est constituée de 2 étapes; Indexation et Recherche
# On va créer une fonction pour l'indexation d'une image
# C'est l'équivalent d'extraire une vecteur descripteur en se basant sur les moments statistiques des couleurs
import numpy as np
def color_moment(image):
  R = image[:, :, 0]
  G = image[:, :, 1]
  B = image[:, :, 2]
  colorFeautures = [np.mean(R),np.std(R),np.mean(G),np.std(G),np.mean(B),np.std(B)]
  colorFeautures /= np.mean(colorFeautures)
  return colorFeautures

def getFeatures(image,fsize):
  return color_moment(image)

def CBIR_Indexation(fsize):
  n= len(loaded_image)
  features = []
  print("Indexation en cours ...")
  for i in range(n):
    features.append(getFeatures(loaded_image[i],fsize))
  print("Indexation terminee. ")
  return features

def Recheche_CBIR(imgReq, ind_Matrix):
    fsize = ind_Matrix[0].shape[0]+1
    imgReq = getFeatures(imgReq, fsize)
    distances = {}
    for i in range(len(ind_Matrix)):
        distances[i] = dist.euclidean(imgReq, ind_Matrix[i])

    distances = sorted([(v, k) for (k, v) in distances.items()])
    return distances

Matrix_indexation = CBIR_Indexation(6)
distances = Recheche_CBIR(image_requete,Matrix_indexation)
print(distances)
fig = plt.figure(figsize=(6, 6), dpi=200)
ax1 = fig.add_subplot(2, 3, 1)
ax1.imshow(image_requete)
ax1.set_title('Image Requete')
#Afficher les 5 premiers simillaires
for i in range(5):
  ax1 = fig.add_subplot(2,3 , i+2)
  plt.imshow(loaded_image[distances[i][1]])
  ax1.set_title(imageNames[distances[i][1]])
plt.show()

```

# Recherche par Histogramme

```python
import cv2 as cv
import numpy as np
from math import ceil

def hsvHistogramFeatures(image):
   [rows,cols,dd] = np.shape(image)
   # convertir l'image RGB en HSV.
   imageHSV = cv.cvtColor(image, cv.COLOR_RGB2HSV)
   #Calcul de l'Histogramme en HSV
   histHSV = cv.calcHist([imageHSV], [0,1,2],None,[8,2,2],[0,180,0,255,0,255])
   hist = histHSV.flatten()
   hist /= np.sum(hist)
   return hist.reshape(-1)

def getFeatures(image,fsize):
  # fonction pour créer le vecteur descripteur
  features = np.zeros((fsize-1 , 1))
  if(fsize >=7):
    features = color_moment(image)
  if(fsize >=39):
    features = np.concatenate((features,hsvHistogramFeatures(image)), axis=None)
  return features

features = CBIR_Indexation(39) #38 est le nombre total des caractéristiques utilisées
```

```python
distanes_CBIR= Recheche_CBIR(image_requete,features)
print(distanes_CBIR)
```

# Recherche par texture

```python
from skimage.feature import graycomatrix, graycoprops
def texture_Feature(image):
  gray = cv.cvtColor(image, cv.COLOR_RGB2GRAY)
  v =[]
  glcm = graycomatrix(gray, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
  v.append(graycoprops(glcm, 'contrast')[0][0])
  v.append(graycoprops(glcm, 'correlation')[0][0])
  v.append(graycoprops(glcm, 'energy')[0][0])
  v.append(graycoprops(glcm, 'homogeneity')[0][0])
  v /= np.sum(v)
  return v

def getFeatures(image,fsize):
  features = np.zeros((fsize-1,1))
  if (fsize >=7):
    features = color_moment(image)
  if (fsize>= 39):
    features = np.concatenate((features,hsvHistogramFeatures(image)),axis=None)
  if (fsize>=43):
    features = np.concatenate((features, texture_Feature(image)),axis=None)
  return features

features = CBIR_Indexation(43)
np.array(features).shape
distances_CBIR= Recheche_CBIR(image_requete,features)
print(distances_CBIR)
```

```python
fig = plt.figure(figsize=(6, 6), dpi=200)
ax1 = fig.add_subplot(2, 3, 1)
ax1.imshow(image_requete)
ax1.set_title('Image Requete')
#Afficher les 5 premiers simillaires
for i in range(5):
  ax1 = fig.add_subplot(2,3 , i+2)
  plt.imshow(loaded_image[distanes_CBIR[i][1]])
  ax1.set_title(imageNames[distanes_CBIR[i][1]])
plt.show()
```

# Recherche par Forme

```python
def shapeFeatures(img):
  image = cv.cvtColor(img,cv.COLOR_RGB2GRAY)
  shapeFeat = cv.HuMoments(cv.moments(image)).flatten()
  shapeFeat /=np.mean(shapeFeat)
  return shapeFeat

def getFeatures(image,fsize):
  features = np.zeros((fsize-1,1))
  if (fsize >=7):
    features = color_moment(image)
  if (fsize>= 39):
    features = np.concatenate((features,hsvHistogramFeatures(image)),axis=None)
  if (fsize>=43):
    features = np.concatenate((features, texture_Feature(image)),axis=None)
  if (fsize >= 50):
    features = np.concatenate((features ,shapeFeatures(image)),axis= None)
  return features

features = CBIR_Indexation(50)
np.array(features).shape
```

```python
distances_CBIR = Recheche_CBIR(image_requete,features)
print(distances_CBIR)
```

```python
fig = plt.figure(figsize=(6, 6), dpi=200)
ax1 = fig.add_subplot(2, 3, 1)
ax1.imshow(image_requete)
ax1.set_title('Image Requete')
#Afficher les 5 premiers simillaires
for i in range(5):
  ax1 = fig.add_subplot(2,3 , i+2)
  plt.imshow(loaded_image[distanes_CBIR[i][1]])
  ax1.set_title(imageNames[distanes_CBIR[i][1]])
plt.show()
```
